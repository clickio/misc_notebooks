```python
#importing algorithms
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
#Split data
df = df.drop(['id'],1)
X = df.drop(['diagnosis'],1).values
y = df['diagnosis'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,test_size=0.2)
```

```python
#Logistic regression
logit = LogisticRegression(max_iter = 3000)
logit.fit(X_test,y_test)
print(cross_val_score(logit,X,y))
```

```python
#use PCAFilter algorithm to filter our data
from kydavra import PCAFilter

max_acc = 0
#loop to choose the best features_number to reduce data
for i in range(1,len(df.columns)-1):
    #creating filter and changing data
    filt = PCAFilter(i)
    df_copy = filt.filter(df,'diagnosis')
    #split the data
    X = df_copy.drop(['diagnosis'],1).values
    y = df_copy['diagnosis'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,test_size=0.2)
    #testing algorithm
    logit = LogisticRegression(max_iter = 1000)
    logit.fit(X_test,y_test)
    y_pred = logit.predict(X_test)
    acc = accuracy_score(y_test,y_pred)
    #saving best features_number
    if(acc>max_acc):
        max_acc = acc
        best_nr = i
print(best_nr,max_acc)

#setting finall data
filt = PCAFilter(best_nr)
df = filt.filter(df,'diagnosis')
```

```python
#Using kydavra MUSESelector to select only important features
from kydavra import MUSESelector

max_acc = 0
for i in range(1,len(df.columns)): 
    #setting features number to i and getting columns
    muse = MUSESelector(num_features=i)
    selected_cols = muse.select(df,'diagnosis')
    #splitting data
    X = df[selected_cols].values
    y = df['diagnosis'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,test_size=0.2)
    #testing algorithm
    logit = LogisticRegression(max_iter = 1000)
    logit.fit(X_test,y_test)
    y_pred = logit.predict(X_test)
    #compare the result to the maximum
    acc = accuracy_score(y_test,y_pred)
    if(acc>max_acc):
        max_acc = acc
        best_nr = i
print(best_nr,max_acc)

#setting finall selector
muse = MUSESelector(num_features=best_nr)
selected_cols = muse.select(df,'diagnosis')
print(selected_cols)
#splitting data
X = df[selected_cols].values
y = df['diagnosis'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,test_size=0.2)
#testing algorithm
logit = LogisticRegression(max_iter = 1000)
logit.fit(X_test,y_test)
y_pred = logit.predict(X_test)
#compare the result to the maximum
acc = accuracy_score(y_test,y_pred)

#creating confusion_matrix
cf = confusion_matrix(y_pred,y_test)
sns.heatmap(cf,annot=True)
```
